# Merging Canada airport and US airport data


```python
# import dependencies
import pandas as pd 
```

```python
# Read in the Canada airport data
canada_airport_data = pd.read_csv("Resources/raw_data/airport_data/airport_data_CA.csv")
canada_airport_data.head()
```

```python
# Read in the US airport data
us_airport_data = pd.read_csv("Resources/raw_data/airport_data/airport_data_US.csv")
us_airport_data.head()
```

```python
# concat canada and us airport data
merged_canada_us_airport_data = pd.concat([canada_airport_data, us_airport_data])
merged_canada_us_airport_data.head()
```

# Filtering the merged airport data to get the Canadian and US 5 busiest airport

```python

#Filter data to get Canadian and 5 businest airport of the USA.
can_us_busiest_airport = merged_canada_us_airport_data.loc[merged_canada_us_airport_data['IATA Code'].isin(["YUL", "YYZ", "YTZ", "YOW", "YKF", "ATL", "DFW", "DEN", "ORD", "JFK"])]
filtered_airports = can_us_busiest_airport.reset_index(drop=True) 
filtered_airports
```

# Read flight price data

```python
# Read in April flight price data
flight_price_data = pd.read_csv("Resources/raw_data/flight_price_data/flight_data_full_April.csv")
flight_price_data.head()
```

# Retrieve unique airports from origin airport and destination airport from flight price dataset

```python
# Get all unique airports  
unique_airports = pd.Series(list(set(flight_price_data['origin_airport']) | set(flight_price_data['destination_airport'])))  # Union of both
unique_airports
```

# Select Canadian and US top 5 busiest airports

```python
# Select only specific airports data in the flight_price dataset
selected_airports = {"Toronto Island"," Kitchener / Waterloo","Ottawa International",
                     "Toronto Pearson International", "Chicago O'Hare International",
                     "Dallas Fort Worth International","Denver International",
                     "Atlanta Hartsfield-Jackson","New York John F. Kennedy",
                     "Montreal Pierre Elliott Trudeau"
                     }
selected_flight_price_data = flight_price_data[flight_price_data['origin_airport'].isin(selected_airports) & flight_price_data['destination_airport'].isin(selected_airports)]
selected_flight_price_data
```

# Map IATA codes to the selected airports in flight price dataset

```python
iata_dict={
    "Toronto Island":"YTZ",
    " Kitchener / Waterloo":"YKF",
    "Ottawa International":"YOW",
    "Toronto Pearson International":"YYZ",
    "Chicago O'Hare International":"ORD",
    "Dallas Fort Worth International":"DFW",
    "Denver International":"DEN",
    "Atlanta Hartsfield-Jackson":"ATL",
    "New York John F. Kennedy":"JFK",
    "Montreal Pierre Elliott Trudeau":"YUL" 
}   

# Map IATA code to airport name
selected_flight_price_data.loc[:, 'origin_airport'] = selected_flight_price_data['origin_airport'].map(iata_dict)
selected_flight_price_data.loc[:, 'destination_airport'] = selected_flight_price_data['destination_airport'].map(iata_dict)
```

```python
selected_flight_price_data
```

```python
selected_flight_price_data = selected_flight_price_data.copy()  # Ensure a full copy
selected_flight_price_data.reset_index(drop=True, inplace=True)
```

```python
selected_flight_price_data.columns
```

# Dropping unwanted columns

```python

columns_to_drop = [
    "sort_by", "price_raw", "origin_city", "origin_country", 
    "destination_city", "destination_country", "change_allowed", 
    "cancellation_allowed", "has_flexible_options","date"
]
selected_flight_price_data = selected_flight_price_data.drop(columns=columns_to_drop, errors="ignore")
selected_flight_price_data.reset_index(drop=True, inplace=True)
selected_flight_price_data
```

# Exploratory Data Analysis

```python
selected_flight_price_data.info()
```

# Renaming the columns

```python
selected_flight_price_data = selected_flight_price_data.rename(columns={
    "origin_airport": "Origin Airport",
    "destination_airport": "Destination Airport",
    "price": "Price",
    "flight_number": "Flight Number",
    "stop_count": "Number of Stops",
    "airline": "Airline",
    "score": "Score"
})
```

```python
selected_flight_price_data
```

```python
#splitting the arrival_time column into two columns arrival_date and arrival_time_only
# Convert arrival_time to datetime format
selected_flight_price_data['departure_time'] = pd.to_datetime(selected_flight_price_data['departure_time'])

# Extract date and time into separate columns arrival_date and arrival_time
selected_flight_price_data['Departure Date'] = selected_flight_price_data['departure_time'].dt.date
selected_flight_price_data['Departure Time'] = selected_flight_price_data['departure_time'].dt.time
selected_flight_price_data
```

```python
#splitting the arrival_time column into two columns arrival_date and arrival_time_only
# Convert arrival_time to datetime format
selected_flight_price_data['arrival_time'] = pd.to_datetime(selected_flight_price_data['arrival_time'])

# Extract date and time into separate columns arrival_date and arrival_time
selected_flight_price_data['Arrival Date'] = selected_flight_price_data['arrival_time'].dt.date
selected_flight_price_data['Arrival Time'] = selected_flight_price_data['arrival_time'].dt.time
selected_flight_price_data
```

```python
# Drop the original column
selected_flight_price_data = selected_flight_price_data.drop(columns=['departure_time'], axis=1)
selected_flight_price_data = selected_flight_price_data.drop(columns=['arrival_time'], axis=1)
```
