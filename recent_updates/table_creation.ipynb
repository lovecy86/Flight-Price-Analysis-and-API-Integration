# Merging cabin_class with flight_price_data

```python
class_flight_price_df = pd.merge(unique_cabin_classes_df,flight_price_data, on='cabin_class')
class_flight_price_df
```

```python
class_flight_price_df= class_flight_price_df.drop(columns='cabin_class', errors="ignore")           
class_flight_price_df
```

# Merging  flight_number with class_flight_price

```python
flight_number_df= pd.merge(class_flight_price_df,unique_flights_df, on='flight_number')
flight_number_df
```

```python
flight_number_df= flight_number_df.drop(columns='flight_number', errors="ignore")           
flight_number_df
```

```python
itinerary_number_df= pd.merge(flight_number_df,itinerary_df, left_on='itinerary_id', right_on='itineraries').rename(columns={'itinerary_id_y':'itinerary_id'})
itinerary_number_df
```

```python
itinerary_number_df= itinerary_number_df.drop(columns=['itinerary_id_x','itineraries'], errors="ignore")           
itinerary_number_df
```

```python
# Merge for origin_airport_id
airport_merged_df = itinerary_number_df.merge(
    airports_df[['airport_id', 'iata code']], 
    left_on='origin_airport', 
    right_on='iata code', 
    how='left'
).rename(columns={'airport_id': 'origin_airport_id'})

# Drop redundant column
airport_merged_df.drop(columns=['iata code'], inplace=True)
airport_merged_df

```

```python
# Merge for destination_airport_id 
airport_merged_df = airport_merged_df.merge(
    airports_df[['airport_id', 'iata code']], 
    left_on='destination_airport', 
    right_on='iata code', 
    how='left'
).rename(columns={'airport_id': 'destination_airport_id'})

# Drop redundant column
airport_merged_df.drop(columns=['iata code'], inplace=True)

# Display final DataFrame
airport_merged_df
```

```python
# Merge for operating_airline_id
airline_merged_df = airport_merged_df.merge(
    airlines_df[['airline_id', 'airline_name']], 
    left_on='operating_airline', 
    right_on='airline_name',
    how='left'
).rename(columns={'airline_id': 'operating_airline_id'})

# Drop redundant column
airline_merged_df.drop(columns=['airline_name'], inplace=True)

# Merge for marketing_airline_id **without overwriting airline_merged_df**
airline_merged_df = airline_merged_df.merge(
    airlines_df[['airline_id', 'airline_name']], 
    left_on='marketing_airline', 
    right_on='airline_name',
    how='left',
    suffixes=('', '_marketing')  # Prevent overwriting columns
).rename(columns={'airline_id': 'marketing_airline_id'})

# Drop redundant column
airline_merged_df.drop(columns=['airline_name','origin_airport','destination_airport','marketing_airline','operating_airline'], inplace=True)

# Display final DataFrame
airline_merged_df
```

```python
airline_merged_df.columns
```

```python
mapped_flight_data = airline_merged_df[['itinerary_id','flight_id','origin_airport_id','destination_airport_id','departure_date_time','arrival_date_time','one_way_price_candollar','cabin_class_id',
                                        'stop_count','duration_in_hours','is_self_transfer','operating_airline_id','marketing_airline_id','score']]
```
